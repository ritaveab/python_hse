```python
text = "Цены: 100 руб, $50, 75.50 евро, 200₽, 300"

# Задача: Найти все цены с валютами (руб, $, евро, ₽)
# Ожидаемый результат: ['100 руб', '$50', '75.50 евро', '200₽']

import re

text = "Цены: 100 руб, $50, 75.50 евро, 200₽, 300"

pattern = r'\d+(?:\.\d+)?\s*(?:руб|евро)|[$₽]\d+(?:\.\d+)?|\d+(?:\.\d+)?\s*[₽]'

prices = re.findall(pattern, text)
print(prices)
```

```python
complex_text = """
Клиент Иван Иванов (ivan@mail.com, +7-912-345-67-89)
заказал товар #12345 за 1,500.50 руб.
Статус: ВЫПОЛНЕНО. Версия заказа v2.1.5.
Сайт: https://store.com/order/12345
"""
# Задачи:
# 1. Найти email и телефон
# 2. Найти номер заказа (#12345)
# 3. Найти цену (1,500.50 руб)
# 4. Найти версию заказа (v2.1.5)

import re

complex_text = """
Клиент Иван Иванов (ivan@mail.com, +7-912-345-67-89)
заказал товар #12345 за 1,500.50 руб.
Статус: ВЫПОЛНЕНО. Версия заказа v2.1.5.
Сайт: https://store.com/order/12345
"""

print("Исходный текст:")
print(complex_text)
print("\n" + "="*50 + "\n")

# 1. Найти email и телефон
email = re.findall(r'\b[\w.-]+@[\w.-]+\.\w+\b', complex_text)
phone = re.findall(r'\+\d-\d{3}-\d{3}-\d{2}-\d{2}', complex_text)

print("1. Email и телефон:")
print(f"   Email: {email}")
print(f"   Телефон: {phone}")

# 2. Найти номер заказа (#12345)
order_number = re.findall(r'#\d+', complex_text)
print(f"\n2. Номер заказа: {order_number}")

# 3. Найти цену (1,500.50 руб)
price = re.findall(r'\b[\d,]+\.\d+\s*руб\b', complex_text)
print(f"3. Цена: {price}")

# 4. Найти версию заказа (v2.1.5)
version = re.findall(r'v\d+(?:\.\d+)+', complex_text)
print(f"4. Версия заказа: {version}")
```

```python
text = "Привет! Как дела?"
banned_words = ["мат", "оскорбление", "спам", "реклама"]

# Приводим текст к нижнему регистру для проверки запрещенных слов
text_lower = text.lower()

# Проверки (часть уже написана)
contains_banned_words = any(word in text_lower for word in banned_words)
is_english_only = all(ord(c) < 128 for c in text if c != ' ')

# ЗАДАНИЕ: Соберите финальную проверку
# Текст валиден если:
# - длина от 1 до 1000 символов
# - И не содержит запрещенных слов
# - И НЕ состоит только из английских символов

is_valid_length = 1 <= len(text) <= 1000
is_valid = is_valid_length and not contains_banned_words and not is_english_only

# Проверим результат
print(f"Текст: '{text}'")
print(f"Длина текста: {len(text)} символов")
print(f"Валидная длина: {is_valid_length}")
print(f"Содержит запрещенные слова: {contains_banned_words}")
print(f"Только английские символы: {is_english_only}")
print(f"Текст валиден: {is_valid}")
```
