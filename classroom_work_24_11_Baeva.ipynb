```python
### task 1
password = input('Введите пароль: ')
while len(password) < 8:
    print('Пароль слишком короткий. Попробуйте снова.')
    password = input('Введите пароль: ')
if len(password) >= 8:
    print('Пароль принят!')
```

```python
### task 1
# оптимальное решение

password = str()
while len(password) < 8:
  password = str(input())
  print("Пароль слишком короткий")
  if len(password) >= 8:
    print ("Пароль принят")
```

```python
### task 2

commands = ('start', 'stop', 'pause', 'quit')

print("Доступные команды: start, stop, pause, quit")

while True:
    user_command = input("Введите команду: ")

    if user_command in commands:

        if user_command == 'quit':
            print("Выход из программы.")
            break
        else:

            print(f"Выполняю команду: {user_command}")

    else:

        print("Неизвестная команда.")
```

```python
### task 3


shopping_list = []

print("Добро пожаловать в список покупок!")

while True:
    print("\nМеню:")
    print("1) Добавить товар")
    print("2) Удалить товар")
    print("3) Показать список")
    print("4) Выйти")

    choice = input("Выберите действие (1-4): ")

    if choice == "1":
        item = input("Введите товар для добавления: ")
        shopping_list.append(item)
        print(f"Товар '{item}' добавлен в список!")

    elif choice == "2":
        if shopping_list:
            print("Текущий список:", shopping_list)
            item = input("Введите товар для удаления: ")
            if item in shopping_list:
                shopping_list.remove(item)
                print(f"Товар '{item}' удален из списка!")
            else:
                print(f"Товар '{item}' не найден в списке!")
        else:
            print("Список покупок пуст!")

    elif choice == "3":
        if shopping_list:
            print("\nВаш список покупок:")
            for i, item in enumerate(shopping_list, 1):
                print(f"{i}. {item}")
        else:
            print("Список покупок пуст!")

    elif choice == "4":
        print("До свидания!")
        break

    else:
        print("Неверный выбор! Пожалуйста, выберите от 1 до 4.")
```

```python
### task 3
# оптимальное решение

shopping_list = []

while True:
    print('Выберите действие:')
    print('1) Добавить товар')
    print('2) Удалить товар')
    print('3) Показать список')
    print('4) Выйти')

    choice = input('Ваш выбор: ')

    if choice == '1':
        shopping_list.append(input('Введите название товара: '))
    elif choice == '2':
        shopping_list.remove(input('Введите название товара: '))
    elif choice == '3':
        print(shopping_list)
    elif choice == '4':
        print('Выход из программы')
        break
```

```python
### task 4

email_set = set()
while True:
  email_address = input("Введите адрес ")
  email_set.add (email_address)
  if email_address == "":
    print(f"{sorted(email_set)}")
    break
```

"""
sorted:
- применяется к итерируемым типам данных (tuple, list)
- создаёт новый объект, который нужно сохранить в новую переменную
- менее "рискованный", т.к. не меняет исходный объект
"""
my_list = [4, 2, 1, 10]
new_list = sorted(my_list)

"""
sort():
- может применяться только к спискам
- изменяет объект (именно поэтому не может использоваться с кортежем)
- не требует создания новой переменной (шустрее!) и ничего не возвращает
"""
my_list = [4, 2, 1, 10]
my_list.sort()
