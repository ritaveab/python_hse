## ЗАДАНИЕ 1: Базовые графики matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

# Создаем фигуру с 4 subplots (2x2)
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. Линейный график y = x^2 * sin(x)
ax = axes[0, 0]
x = np.linspace(0, 10, 100)
y = x**2 * np.sin(x)
ax.plot(x, y, 'b-', linewidth=2, label='y = x²·sin(x)')
ax.set_title('Линейный график: y = x²·sin(x)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True, alpha=0.3)
ax.legend()

# 2. Гистограмма нормального распределения
ax = axes[0, 1]
data = np.random.normal(0, 1, 1000)
ax.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
ax.set_title('Гистограмма нормального распределения (n=1000)')
ax.set_xlabel('Значение')
ax.set_ylabel('Частота')
ax.grid(True, alpha=0.3)

# 3. Scatter plot: y = 2x + шум
ax = axes[1, 0]
np.random.seed(42)  # Для воспроизводимости результатов
x = np.random.uniform(0, 10, 50)
y = 2 * x + np.random.normal(0, 1, 50)
colors = np.random.rand(50)  # Случайные цвета
sizes = np.random.uniform(30, 100, 50)  # Случайные размеры
ax.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis', edgecolors='black', linewidth=0.5)
ax.set_title('Scatter plot: y = 2x + шум (n=50)')
ax.set_xlabel('x (равномерное распределение)')
ax.set_ylabel('y = 2x + шум')
ax.grid(True, alpha=0.3)

# 4. Bar chart: точность моделей
ax = axes[1, 1]
models = ['A', 'B', 'C', 'D', 'E']
accuracy = [0.85, 0.92, 0.88, 0.95, 0.90]
bars = ax.bar(models, accuracy, color=['#4CAF50', '#2196F3', '#FF9800', '#9C27B0', '#F44336'])
ax.set_title('Сравнение точности моделей')
ax.set_xlabel('Модель')
ax.set_ylabel('Точность')
ax.set_ylim(0.8, 1.0)
ax.grid(True, alpha=0.3, axis='y')

# Добавляем значения над столбцами
for bar, acc in zip(bars, accuracy):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height + 0.005,
            f'{acc:.2f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# Используйте датасет tips из seaborn:
# import seaborn as sns
# tips = sns.load_dataset('tips')

# Создайте 3 графика:
# 1. Box plot: total_bill по дням недели (day)
#    Разделите дополнительно по времени (time) с помощью hue
# 2. Violin plot: tip в зависимости от размера компании (size)
# 3. Pair plot: выберите 4 числовых признака, раскрасьте по полу (sex)

import seaborn as sns
import matplotlib.pyplot as plt

# Загружаем датасет
tips = sns.load_dataset('tips')

# Создаем фигуру
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# 1. Box plot: total_bill по дням с разделением по времени
sns.boxplot(data=tips, x='day', y='total_bill', hue='time', ax=axes[0],
            palette='Set2', linewidth=1.5, fliersize=4)
axes[0].set_title('Распределение суммы счета по дням и времени', fontsize=14, fontweight='bold')
axes[0].set_xlabel('День недели', fontsize=12)
axes[0].set_ylabel('Сумма счета ($)', fontsize=12)
axes[0].legend(title='Время', loc='upper right')
axes[0].grid(True, alpha=0.3, axis='y')

# 2. Violin plot: tip по размеру компании
sns.violinplot(data=tips, x='size', y='tip', ax=axes[1],
               palette='viridis', linewidth=1.5, inner='quartile')
axes[1].set_title('Распределение чаевых по размеру компании', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Размер компании', fontsize=12)
axes[1].set_ylabel('Чаевые ($)', fontsize=12)
axes[1].grid(True, alpha=0.3, axis='y')

# Добавляем значения медиан на violin plot
for i, size in enumerate(sorted(tips['size'].unique())):
    median_tip = tips[tips['size'] == size]['tip'].median()
    axes[1].text(i, median_tip, f'{median_tip:.2f}',
                 ha='center', va='bottom', fontweight='bold', fontsize=10)

plt.tight_layout()
plt.show()

# 3. Pair plot: 4 числовых признака, раскраска по полу
print("\n" + "="*70)
print("Pair plot: взаимосвязи между признаками (раскраска по полу)")
print("="*70)

# Создаем отдельную фигуру для pair plot
pair_grid = sns.pairplot(
    data=tips[['total_bill', 'tip', 'size', 'sex']],
    hue='sex',
    palette={'Male': '#1f77b4', 'Female': '#ff7f0e'},
    diag_kind='kde',  # Используем kde вместо гистограммы для диагонали
    plot_kws={'alpha': 0.7, 's': 50, 'edgecolor': 'black'},
    height=2.5
)
```
